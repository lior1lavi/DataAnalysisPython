```python
import pandas as pd
import seaborn
```

```python
'The seaborn package includes some data sets that we can use'
flights = seaborn.load_dataset('flights')
```

```python
'Using the HEAD function, we can present only the begining of the data stet'
flights.head()
```

```python
flights_indexed = flights.set_index(['year','month'])
flights_indexed.head()
```

```python
flights_indexed.loc[1949]
```

```python
'Define two column as index and select a range of years'
flights_indexed.loc[1949:1950]
```

```python
'Selecg specific year and month using he LOC function'
flights_indexed.loc[1949,'January']
```

```python
'Selecg specific year and range of months using he LOC function'
flights_indexed.loc[1949].loc['January':'June']
```

```python
'Define rows as columns using the UNSTACKED function'
flights_unstacked = flights_indexed.unstack()
```

```python
flights_unstacked
```

```python
'We can now sum all the rows in the year column
flights_unstacked.sum(axis=1)
```

```python
'We can assign the results to a new column and now the name of the column become compund labels
flights_unstacked['passengers','total'] = flights_unstacked.sum(axis=1)
```

```python
flights_unstacked
```

```python
flights_restacked = flights_unstacked.stack()
```

```python
flights_restacked
```

```python
flights_restacked.loc[pd.IndexSlice[:,'total'],'passengers']
```

```python
flights_restacked[flights_restacked['passengers'] > 120]
```
