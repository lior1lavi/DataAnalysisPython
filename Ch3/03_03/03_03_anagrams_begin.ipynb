```python
'Find anagrams for all the words in the dictionary file'
'1: Create a python dictionary of all the words, idexed by the signature of the word in the list'
'2: In every item in the dictionary we will have the sugrnature as the key and a list of words as the value'
'3: We will loop throw a list of words and then append to the right item with the signature as the key'
'We must ensure that all the items (words) exists in the dictionary using the collection module'
'4: Using a default dictionary object to give a default value for item that does not exists'
'The default value we will use is an empty list'
import collections

words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
words_bysig
```

```python
'Define function which finds anagram for a specific word'
'Using a simple dictionary lookup'
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagram_fast('dictionary')
```

```python
'Define function to find all the anagrams in the dictionary'
'1: WE will build this as a dictionary comprehantion '
'2: The key are the word to be anagrams'
'3: The values are the list of co responding anagrams'
'4: Excluding the duplicates words (lookup word to itself)'
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```
