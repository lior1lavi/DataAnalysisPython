```python
def parsefile(filename):
    return np.genfromtxt(filename,
                         delimiter = dly_delimiter,
                         usecols = dly_usecols,
                         dtype = dly_dtype,
                         names = dly_names)
```

```python
dly_delimiter = [11,4,2,4] + [5,1,1,1] * 31
dly_usecols = [1,2,3] + [4*i for i in range(1,32)]
dly_dtype = [np.int32,np.int32,(np.str_,4)] + [np.int32] * 31
dly_names = ['year','month','obs'] + [str(day) for day in range(1,31+1)]
```

```python
def unroll(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'],record['month']))
    dates = np.arange(startdate,startdate + np.timedelta64(1,'M'),np.timedelta64(1,'D'))
    
    rows = [(date,record[str(i+1)]/10) for i,date in enumerate(dates)]
    
    return np.array(rows,dtype=[('date','M8[D]'),('value','d')])
```

```python
def getobs(filename,obs):
    return np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
```

```python
def getobs(filename,obs):
    data = np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
    data['value'][data['value'] == -999.9] = np.nan
    
    return data
```

```python
'Since there are allot of noise in the data, we will mean the data points'
'It will be avaraging the above values'
'This will include mean over a limited windows centered at the data points'
'We can dial up or down the size of the window'
'We will define aplot smooth function that gets the size of the window'
'The numpy function "correlate" mutilply the sliding section of a one dimention array with another shorter array'
'Compute the sum of the arrays and stors it in a new array'
def plot_smoothed(t,win = 10):
    smoothed = np.correlate(t['value'], np.ones(win)/win,'same')
    
    pp.plot(t['date'],smoothed)
```

```python
plot_smoothed(lihue_tmin)
```

```python
'Lets look at the smallest slice of this array'
'Also, lets compare it with aslot without smoothing'
pp.plot(lihue_tmin[10000:12000]['date'],lihue_tmin[10000:12000]['value'])

plot_smoothed(lihue_tmin[10000:12000])
plot_smoothed(lihue_tmin[10000:12000],30)
```

```python
'Lets plot all of the stations together into asmoothing operation'
'We keep the code of the stations in the list "datastations"'
'We will call the smooth function on the array of each station using the file name calling the TMIN/TMAX values'
'Adding 365 as the smoothing parameter window size'
'Also, we need to tell matplotlib to palce this plot in different sub plot within a larger one using the "subplot" command'
'Using the "figure" function, we can design the size of the graph'
'Finaly, we will plot the same range of year/temp for all the stations using the "axis" function'
'Last, we can tell matplotlib to applay a "tight_layout" function and give a title for the legends using the "title()" function'
    
pp.figure(figsize=(10,6))    
for i,code in enumerate(datastations):
    pp.subplot(2,2,i+1)
    
    plot_smoothed(getobs('{}.dly'.format(code),'TMIN'),365)
    plot_smoothed(getobs('{}.dly'.format(code),'TMAX'),365)
    
    pp.title(stations[code])
    pp.axis(xmin=np.datetime64('1952'),xmax=np.datetime64('2012'),ymin = -10, ymax = 30)
    
pp.tight_layout()
```
